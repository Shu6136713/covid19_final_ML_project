see that the format is 1 true, 2 false, 99/98/97 unknown. let fix the df

```python
def fix_unknown(df: pd.DataFrame, columns:list):
    for col in columns:
        df[col] = [1 if val==1 else 0 if val==2 else np.nan for val in df[col]]
    return df
```

```python
show_df = df.copy()
```

```python
def label_group(row):
    if row['PNEUMONIA'] == 1 and row['COPD'] == 1 and row['ASTHMA']:
        return 'both'
    elif row['PNEUMONIA'] == 1:
        return 'pneumonia only'
    elif row['COPD'] == 1:
        return 'chronic only'
    elif row['ASTHMA']:
        return "astma only"
    else:
        return 'neither'

show_df['lung_condition_group'] = show_df.apply(label_group, axis=1)
```

```python
sns.histplot(show_df, x="lung_condition_group", hue="DEATH", multiple="dodge", shrink=0.8 )
```

```python
death_rates = show_df.groupby("lung_condition_group")["DEATH"].agg(['sum', 'count'])
death_rates["death_rate"] = death_rates["sum"] / death_rates["count"]
death_rates = death_rates.reset_index()
```
